```python
data_ = data.head(400)
```

```python
import networkx as nx

# Example: Create edges based on common engagement metrics
G = nx.Graph()

for _, row in data_.iterrows():
    G.add_node(row['Tiktok name'],
               subscribers=row['Subscribers'],
               views_avg=row['Views avg.'],
               likes_avg=row['Likes avg.'],
               comments_avg=row['Comments avg.'],
               shares_avg=row['Shares avg.'])
```

```python
degree_centrality = nx.degree_centrality(G)
nx.set_node_attributes(G, degree_centrality, 'degree_centrality')
```

```python
influencers = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
top_10_influencers = influencers[:10]
print("Top 10 Influencers:", top_10_influencers)
```

```python
import matplotlib.pyplot as plt

# Layout for visualization
pos = nx.spring_layout(G)  # Force-directed layout for better spacing

missing_nodes = [node for node in G.nodes if 'weighted_degree_centrality' not in G.nodes[node]]
print("Nodes missing centrality attribute:", missing_nodes)
```

```python
for node in G.nodes:
    if 'weighted_degree_centrality' not in G.nodes[node]:
        G.nodes[node]['weighted_degree_centrality'] = 0
```

```python
node_sizes = [5000 * G.nodes[node]['weighted_degree_centrality'] for node in G.nodes]
```

```python
plt.figure(figsize=(20, 15))
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='blue', alpha=0.7)
nx.draw_networkx_edges(G, pos, alpha=0.5, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')
plt.title("TikToker Network Visualization")
plt.show()
```
